```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('postgresql+psycopg2://mgpu_ico_st_09:KkvE7mrh@95.131.149.21:5432/bi_sql_data_student')

print("📁 СОЗДАНИЕ CSV ФАЙЛОВ ДЛЯ DATALENS:")
print("="*50)

# 1. Данные о сотрудниках
employees_df = pd.read_sql("""
    SELECT
        employee_id,
        first_name || ' ' || last_name AS full_name,
        email,
        role
    FROM pm.employees
    ORDER BY employee_id
""", engine)
employees_df.to_csv("datalens_employees.csv", index=False, encoding='utf-8-sig')
print(f"✅ datalens_employees.csv - {len(employees_df)} сотрудников")

# 2. Данные о проектах
projects_df = pd.read_sql("""
    SELECT
        project_id,
        project_name,
        start_date,
        end_date,
        status,
        CASE
            WHEN end_date IS NULL THEN 'В работе'
            WHEN end_date < CURRENT_DATE THEN 'Завершен'
            ELSE 'Активен'
        END AS current_state
    FROM pm.projects
    ORDER BY project_id
""", engine)
projects_df.to_csv("datalens_projects.csv", index=False, encoding='utf-8-sig')
print(f"✅ datalens_projects.csv - {len(projects_df)} проектов")

# 3. Данные о задачах
tasks_df = pd.read_sql("""
    SELECT
        t.task_id,
        t.title AS task_name,
        p.project_name,
        t.due_date,
        t.priority,
        CASE
            WHEN t.priority = 1 THEN 'Высокий'
            WHEN t.priority = 2 THEN 'Средний'
            ELSE 'Низкий'
        END AS priority_name,
        CASE
            WHEN t.due_date < CURRENT_DATE THEN 'Просрочена'
            WHEN t.due_date = CURRENT_DATE THEN 'Сегодня'
            ELSE 'В работе'
        END AS task_status
    FROM pm.tasks t
    JOIN pm.projects p ON t.project_id = p.project_id
    ORDER BY p.project_name, t.priority
""", engine)
tasks_df.to_csv("datalens_tasks.csv", index=False, encoding='utf-8-sig')
print(f"✅ datalens_tasks.csv - {len(tasks_df)} задач")

# 4. Данные о назначениях (сотрудники + задачи)
assignments_df = pd.read_sql("""
    SELECT
        a.assignment_id,
        e.first_name || ' ' || e.last_name AS employee_name,
        e.role AS employee_role,
        t.title AS task_name,
        p.project_name,
        a.assigned_date
    FROM pm.assignments a
    JOIN pm.employees e ON a.employee_id = e.employee_id
    JOIN pm.tasks t ON a.task_id = t.task_id
    JOIN pm.projects p ON t.project_id = p.project_id
    ORDER BY p.project_name, e.last_name
""", engine)
assignments_df.to_csv("datalens_assignments.csv", index=False, encoding='utf-8-sig')
print(f"✅ datalens_assignments.csv - {len(assignments_df)} назначений")

# 5. Сводная таблица для дашборда
dashboard_df = pd.read_sql("""
    SELECT
        p.project_name,
        p.status AS project_status,
        COUNT(DISTINCT t.task_id) AS total_tasks,
        COUNT(DISTINCT a.employee_id) AS assigned_employees,
        COUNT(CASE WHEN t.priority = 1 THEN 1 END) AS high_priority_tasks,
        COUNT(CASE WHEN t.priority = 2 THEN 1 END) AS medium_priority_tasks,
        COUNT(CASE WHEN t.priority = 3 THEN 1 END) AS low_priority_tasks,
        MIN(t.due_date) AS earliest_deadline,
        MAX(t.due_date) AS latest_deadline
    FROM pm.projects p
    LEFT JOIN pm.tasks t ON p.project_id = t.project_id
    LEFT JOIN pm.assignments a ON t.task_id = a.task_id
    GROUP BY p.project_id, p.project_name, p.status
    ORDER BY total_tasks DESC
""", engine)
dashboard_df.to_csv("datalens_dashboard.csv", index=False, encoding='utf-8-sig')
print(f"✅ datalens_dashboard.csv - {len(dashboard_df)} проектов")

print("\n" + "="*50)
print("📊 ИТОГОВАЯ СТАТИСТИКА:")
print("="*50)
print(f"👥 datalens_employees.csv - {len(employees_df)} сотрудников")
print(f"📁 datalens_projects.csv - {len(projects_df)} проектов")
print(f"✅ datalens_tasks.csv - {len(tasks_df)} задач")
print(f"🔗 datalens_assignments.csv - {len(assignments_df)} назначений")
print(f"📊 datalens_dashboard.csv - {len(dashboard_df)} проектов (сводка)")

print("\n" + "="*50)
print("✅ ГОТОВО! Теперь у вас 5 CSV файлов для DataLens:")
print("="*50)
print("1. datalens_employees.csv - список сотрудников")
print("2. datalens_projects.csv - список проектов")
print("3. datalens_tasks.csv - задачи с приоритетами")
print("4. datalens_assignments.csv - назначения сотрудников на задачи")
print("5. datalens_dashboard.csv - сводная статистика для дашборда")
```
